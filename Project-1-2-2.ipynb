```python
def preprocessing_1(data_frame):
    df = data_frame.copy()
    df = df.dropna(subset=["literacy"])
    df.loc[df['literacy'] == 'illiterate', 'studying'] = 'No'
    df.loc[df['literacy'] == 'illiterate', 'degree'] = 'without'
    df = df.dropna(subset=["occupationalst"])
    return df
```

```python
## Report cell | is preprocessing_1 worked?
temp_new_R98 = preprocessing_1(raw_data_R98)

df = temp_new_R98

print(1, '*'*81)
for column in columns_1+columns_2:
    print(
        column, 
        df[column].isnull().sum(),
        len(df[column]),
        ( df[column].isnull().sum() * 100 /len(df[column]) ),
        sep = " | "
    )  

print(2, '*'*81)
print(
    len(raw_data_R98) - len(df),
    (len(raw_data_R98) - len(df)) * 100 / len(raw_data_R98),
    sep = " | "
)
```

```python
def degree_encoder(category):
    mapping = {'without': 0, 'Other': 9, 'Elemantry': 1, 'Secondary': 2, 
                   'HighSchool': 3, 'Diploma': 4, 'College': 5, 'Bachelor': 6, 
                   'Master': 7, 'PhD': 8}
    return mapping[category]

main_R98_1["degree_encoded"] = main_R98_1["degree"].apply(degree_encoder)
main_R99_1["degree_encoded"] = main_R99_1["degree"].apply(degree_encoder)
main_R1400_1["degree_encoded"] = main_R1400_1["degree"].apply(degree_encoder)
main_R1401_1["degree_encoded"] = main_R1401_1["degree"].apply(degree_encoder)

main_U98_1["degree_encoded"] = main_U98_1["degree"].apply(degree_encoder)
main_U99_1["degree_encoded"] = main_U99_1["degree"].apply(degree_encoder)
main_U1400_1["degree_encoded"] = main_U1400_1["degree"].apply(degree_encoder)
main_U1401_1["degree_encoded"] = main_U1401_1["degree"].apply(degree_encoder)
```
